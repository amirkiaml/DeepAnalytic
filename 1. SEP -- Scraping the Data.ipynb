# 2. Scrape the Website

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

df = pd.read_parquet('SEP.parquet')

for i in range(0, len(df)):
    url = df.at[i, 'Url']
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        try:
            # Extract text or another attribute instead of the whole tag
            title = soup.find('h1').get_text() if soup.find('h1') else None
            preamble = soup.find('div', id='preamble').get_text() if soup.find('div', id='preamble') else None
            toc = soup.find('div', id='toc').get_text() if soup.find('div', id='toc') else None
            text = soup.find('div', id='main-text').get_text() if soup.find('div', id='main-text') else None
            bib = soup.find('div', id='bibliography').get_text() if soup.find('div', id='bibliography') else None
            other = soup.find('div', id='other-internet-resources').get_text() if soup.find('div', id='other-internet-resources') else None
            related = soup.find('div', id='related-entries').get_text() if soup.find('div', id='related-entries') else None
            copyright = soup.find('div', id='article-copyright').get_text() if soup.find('div', id='article-copyright') else None

            # Assign the text to the DataFrame
            df.at[i, 'Title'] = title
            df.at[i, 'Preamble'] = preamble
            df.at[i, 'TOC'] = toc
            df.at[i, 'Text'] = text
            df.at[i, 'Bib'] = bib
            df.at[i, 'Other Resources'] = other
            df.at[i, 'Related'] = related
            df.at[i, 'Copyright'] = copyright
            
            df.to_parquet('SEP.parquet')

        except AttributeError:
            print("Failed to find some elements in the HTML.")
    else:
        print(f"Failed to retrieve the webpage {url}. Status code: ", response.status_code)
```
